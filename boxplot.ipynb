```python
import pandas as pd
import numpy as np
from sklearn import preprocessing as prep
from collections import Counter
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

input_file = pd.read_csv('Dataset_.csv')
dataframe = input_file.drop(columns=['Day', 'Type'])
print(dataframe)
```

```python
arr = StandardScaler().fit_transform(dataframe)
#arr = RobustScaler().fit_transform(dataframe)
#arr = MinMaxScaler().fit_transform(dataframe)
DF = pd.DataFrame()
arr_outliers_iqr, arr_outliers_595, arr_outliers_298 = [], [], []
i=0
for column in dataframe.columns:
    perc25 = np.percentile(arr, 25)
    perc50 = np.percentile(arr, 50)
    perc75 = np.percentile(arr,75)
    whiskers = 1.5 * (perc75 - perc25)
    perc5 = np.percentile(arr, 5)
    perc95 = np.percentile(arr, 95)
    for j in range(len(arr[:, 0])):
        if arr[j][i] > perc75+whiskers or arr[j][i] < perc25-whiskers:
            arr_outliers_iqr.append(j)
    for k in range(len(arr[0])):
        if arr[k][i] > perc95 or arr[k][i] < perc5:
            arr_outliers_595.append(k)
    for l in range(len(arr[0])):
        if arr[l][i] > np.percentile(arr,98) or arr[l][i] < np.percentile(arr,2):
            arr_outliers_298.append(l)       
    series = pd.Series(arr[:,i])
    DF.insert(i,column,series)
    i+=1
```

```python
arr_outliers_595.sort()
print(DF)
print(Counter(arr_outliers_iqr))
print(Counter(arr_outliers_595))
print(Counter(arr_outliers_298))
plt.figure(figsize=(20,10))
dataframe.boxplot()
plt.show()
plt.figure(figsize=(20,10))
DF.boxplot()
plt.show()
```
